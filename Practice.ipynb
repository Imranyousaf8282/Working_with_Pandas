```python
ask_bol = hn["title"].str.lower().str.startswith("ask hn")
```

```python
ask_bol
```

```python
ask_posts  = hn[ask_bol]
ask_posts
```

```python
show_bol = hn["title"].str.lower().str.startswith("show hn")
```

```python
show_bol
```

```python
show_posts = hn[show_bol]
show_posts
```

```python
show_posts["num_comments"].mean()
```

```python
ask_posts["num_comments"].mean()
```

```python
ask_posts.info()
```

```python
ask_posts["created_at"] = pd.to_datetime(ask_posts["created_at"]).copy()
```

```python
ask_posts["created_at"]
```

```python
ask_posts["created_at"].dt.hour
```

```python
ask_posts["hour"] = ask_posts["created_at"].dt.hour
```

```python
ask_posts
```

```python
avg_comment  = ask_posts.pivot_table(index="hour",values="num_comments",aggfunc="mean")
avg_comment
```

```python
avg_comment.sort_values("num_comments",ascending=False)
```
