## Working With Hacker News Data

```python
import pandas as pd
```

```python
file = "hacks.csv"

hn = pd.read_csv(file)
```

```python
hn.head()
```

```python
hn.tail(50)
```

```python
hn["title"]
```

```python
"ASK HN"
"Ask Hn"
"ask hn"
```

```python
ask_bol = hn["title"].str.lower().str.startswith("ask hn")
ask_bol
```

```python
ask_posts = hn[ask_bol]
ask_posts.head()
```

```python
show_bol = hn["title"].str.lower().str.startswith("show hn")
show_bol
```

```python
show_posts = hn[show_bol]
show_posts.head()
```

```python
ask_posts["num_comments"].mean()
```

```python
show_posts["num_comments"].mean()
```

#### working with ask_posts

```python
ask_posts.info()
```

```python
# first we convert object in to datetime
ask_posts["created_at"] = pd.to_datetime(ask_posts["created_at"])
```

```python
ask_posts["created_at"].head()
```

```python
ask_posts["hour"] = ask_posts["created_at"].dt.hour.copy()
```

```python
ask_posts
```

```python
ask_posts["hour"]
```

```python
avg_comnts = ask_posts.pivot_table(index="hour",values="num_comments",aggfunc= "mean")
```

```python
avg_comnts
```

```python
avg_comnts.sort_values("num_comments")
```

```python
avg_comnts.sort_values("num_comments",ascending = False)
```

```python
avg_comnts.sort_values("num_comments").sum()
avg_comnts.sort_values("num_comments").max()
avg_comnts.sort_values("num_comments").min()
```
