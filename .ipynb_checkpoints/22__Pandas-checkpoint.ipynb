## Working on Date column

```python
arts["Date"].head(20)
```

```python
bad_chr = ["(",")",".","c","C","s","'"," "]

len(bad_chr)
```

```python
date = "(c.1947's)"

for i in bad_chr:
    date = date.replace(i,"")
```

```python
date
```

```python
def clean_date(date):
    bad_chr = ["(",")",".","c","C","s","'"," "]
    for i in bad_chr:
        date = date.replace(i,"")
    return date
    
```

```python
clean_date(date)
```

```python
clean_date("cs.1947's'C")
```

```python
arts["Date"] = arts["Date"].apply(clean_date)
```

```python
arts
```

```python
new_date = "1889-1911"

lst = new_date.split("-")
lst
```

```python
arts["Date"]
```

```python
def int_date(date):
    if "-" in date:
        lst = new_date.split("-")
        yr1 = int(lst[0])
        yr2 = int(lst[1])
        date = round((yr1 + yr2)/2)

    else:
        date = int(date)

    return date
```

```python
    int_date(date)
```

```python
int_date("2999")
```

```python
arts["Date"] = arts["Date"].apply(int_date)
```
