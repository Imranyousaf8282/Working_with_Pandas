```python
def cat_exp(experience_coding):
    if experience_coding < 5:
        return "Less then 5 Years"
    elif experience_coding in range(5,11):
        return "5-10 Years"
    elif experience_coding in range(11,16):
        return "10-15 Years"
    elif experience_coding in range(16,21):
        return "15-20 Years"
    elif experience_coding in range(21,26):
        return "20-25 Years"
    else:
        return "25+ Years"
```

```python
dss.experience_coding.apply(cat_exp)
```

```python
dss["experience_coding_cat"] = dss.experience_coding.apply(cat_exp)
```

```python
dss
```

```python
dss.groupby("experience_coding_cat")["compensation"].mean().round(2)
```
